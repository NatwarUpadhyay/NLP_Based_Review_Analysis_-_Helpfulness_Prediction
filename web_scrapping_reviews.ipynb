```python
link='https://www.flipkart.com/apple-iphone-11-black-128-gb/product-reviews/itm8244e8d955aba?pid=MOBFWQ6BKRYBP5X8&lid=LSTMOBFWQ6BKRYBP5X8X0KYUG&aid=overall&certifiedBuyer=false&sortOrder=MOST_RECENT'

ratings = []
names = []
dates = []
comments = []
reviews = []
likes = []
dislikes = []
noOfImages = []


for pageNo in range(0, 500):
    page = requests.get(link+'&page='+ str(pageNo+1))
    soup = bs(page.content, 'html.parser')
    name=soup.find('div',class_="_27M-vq")
    
    for data in soup.findAll('div',class_='_27M-vq'):
        
   #Get Rating
        review_rating=str(data.find('div', attrs={'class':'_3LWZlK _1BLPMq'}))
        rating_pattern = r'<div class="_3LWZlK _1BLPMq">(\d)<img'
        rating_value = re.search(rating_pattern, review_rating)
        if rating_value:
            rating_value = rating_value.group(1)
        else:
            rating_value = '3'
            
        ratings.append(rating_value)
            
    #Get Review Comment 
        comment=data.find('p', attrs={'class':'_2-N8zT'}).text.strip()
        
        comments.append(comment)
        
    #Get Reviewer Name
        review_name = data.find('p', attrs={'class': '_2sc7ZR _2V5EHH'}).text.strip()
        
        names.append(review_name)
        
    #Get Review Date
        review_date = data.find_all('p', {'class': "_2sc7ZR"})
        for i in range(1, len(review_date), 2):  # Start from the second element, skipping every other one
            review_date = review_date[i].get_text()
        match = re.match(r'(\d+)\s+months? ago', review_date)
        if match:
            months_ago = int(match.group(1))
            # Calculate the actual date
            actual_date = datetime.now() - relativedelta(months=months_ago)
            dates.append(actual_date)
        else:
            dates.append(review_date)
    
    #Get Review
        review = data.find('div', {'class': 't-ZTKy'}).text.strip()
        updated_review = review.split("READ MORE")[0].strip()
        
        reviews.append(updated_review)
    
    #Get Number of Likes
        like = data.find('span', {'class': '_3c3Px5'}).text.strip()
        
        likes.append(like)
        
    #Get Number of Dislikes
        dislike = data.find('div',attrs={'class':'_1LmwT9 pkR4jH'})
        dislike = dislike.find('span', {'class': '_3c3Px5'}).text.strip()
        
        dislikes.append(dislike)

    #Get Number of Images
        review_images = data.find('div',attrs={'class':'_2nMSwX _3oLIki'})
        if review_images != None:
            div_elements = review_images.find_all('div')
            review_no_of_images = len(div_elements)
        else:
            review_no_of_images = '0'

        noOfImages.append(review_no_of_images)
  
        print("************************")
        print('name:'+str(review_name))
        print('Date:'+str(actual_date))
        print('rating:'+str(rating_value))
        print('comment:'+str(comment))
        print('review:'+str(updated_review))
        print('likes:'+str(like))
        print('dislikes:'+str(dislike))
        print('noOfImg:'+str(review_no_of_images))
        print("************************")
        
# # Create a dictionary with the extracted data
data = {'Rating': ratings, 'Name': names, 'Date': dates, 'Comments': comments, 'Reviews': reviews, 'Likes': likes, 'Dislikes': dislikes, 'Number of Images':noOfImages}

print(data)

df = pd.DataFrame(data)

# Save DataFrame to a CSV file
df.to_csv('reviews_1.csv', index=False)
```

```python
# Function to scrape data from a single page
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    review_titles = []

    for review in soup.find_all('div', {'class': '_27M-vq'}):
        review_title = review.find('p', {'class': '_2-N8zT'}).text.strip()

        review_titles.append(review_title)

    # Return data from this page
    return {
        'Review Title': review_titles,
    }
# URL pattern for different pages
url_pattern = "https://www.flipkart.com/apple-iphone-11-black-128-gb/product-reviews/itm8244e8d955aba?pid=MOBFWQ6BKRYBP5X8&lid=LSTMOBFWQ6BKRYBP5X8X0KYUG&aid=overall&certifiedBuyer=false&sortOrder=MOST_HELPFUL&page={}"

# Scrape data from 2 pages for testing
all_data = []
for page_num in range(0, 500):
    url = url_pattern.format(page_num)
    page_data = scrape_page(url)
    all_data.append(page_data)

# Combine data from all pages into a single DataFrame
df = pd.DataFrame({
    'Review Title': sum((data['Review Title'] for data in all_data), []),
})
# Save DataFrame to a CSV file
df.to_csv('review_titles.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
from datetime import datetime
from dateutil.relativedelta import relativedelta

link = 'https://www.flipkart.com/apple-iphone-11-black-128-gb/product-reviews/itm8244e8d955aba?pid=MOBFWQ6BKRYBP5X8&lid=LSTMOBFWQ6BKRYBP5X8X0KYUG&aid=overall&certifiedBuyer=false&sortOrder=MOST_HELPFUL'

ratings = []
names = []
dates = []
comments = []
reviews = []
likes = []
dislikes = []
noOfImages = []
review_titles = []  # Initialize list to store review titles

for pageNo in range(0, 500):
    page = requests.get(link+'&page='+ str(pageNo+1))
    soup = bs(page.content, 'html.parser')
    
    for data in soup.findAll('div', class_='_27M-vq'):
        # Get Rating
        review_rating = str(data.find('div', class_='_3LWZlK _1BLPMq'))
        rating_pattern = r'<div class="_3LWZlK _1BLPMq">(\d)<img'
        rating_value = re.search(rating_pattern, review_rating)
        if rating_value:
            rating_value = rating_value.group(1)
        else:
            rating_value = '3'
        ratings.append(rating_value)

        # Get Review Title
        review_title = data.find('p', class_='_2-N8zT').text.strip()
        review_titles.append(review_title)

        # Get Reviewer Name
        review_name = data.find('p', class_='_2sc7ZR _2V5EHH').text.strip()
        names.append(review_name)

        # Get Review Date
        review_date = data.find_all('p', {'class': "_2sc7ZR"})
        for i in range(1, len(review_date), 2):  # Start from the second element, skipping every other one
            review_date = review_date[i].get_text()
        match = re.match(r'(\d+)\s+months? ago', review_date)
        if match:
            months_ago = int(match.group(1))
            # Calculate the actual date
            actual_date = datetime.now() - relativedelta(months=months_ago)
            dates.append(actual_date)
        else:
            dates.append(review_date)

        # Get Review
        review = data.find('div', class_='t-ZTKy').text.strip()
        updated_review = review.split("READ MORE")[0].strip()
        reviews.append(updated_review)

        # Get Number of Likes
        like = data.find('span', class_='_3c3Px5').text.strip()
        likes.append(like)

        # Get Number of Dislikes
        dislike = data.find('div', class_='_1LmwT9 pkR4jH').find('span', class_='_3c3Px5').text.strip()
        dislikes.append(dislike)

        # Get Number of Images
        review_images = data.find('div', class_='_2nMSwX _3oLIki')
        if review_images:
            div_elements = review_images.find_all('div')
            review_no_of_images = len(div_elements)
        else:
            review_no_of_images = '0'
        noOfImages.append(review_no_of_images)

        # Get Comment
        comment = data.find('p', class_='_2-N8zT').text.strip()
        comments.append(comment)

        print("************************")
        print('Title: ' + review_title)
        print('Name: ' + str(review_name))
        print('Date: ' + str(actual_date))
        print('Rating: ' + str(rating_value))
        print('Comment: ' + str(comment))
        print('Review: ' + str(updated_review))
        print('Likes: ' + str(like))
        print('Dislikes: ' + str(dislike))
        print('No. of Images: ' + str(review_no_of_images))
        print("************************")

# Create a dictionary with the extracted data
data = {'Rating': ratings, 'Name': names, 'Date': dates, 'Comments': comments, 'Reviews': reviews, 'Likes': likes, 'Dislikes': dislikes, 'Number of Images': noOfImages, 'Review Title': review_titles}

df = pd.DataFrame(data)

# Save DataFrame to a CSV file
df.to_csv('reviews_1.csv', index=False)
```
